```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor

# Zufallsdaten generieren (z. B. Hausgröße in sqft und Preis in USD)
np.random.seed(42)
# Hausgrößen zwischen 100 und 220 qm
X = np.random.randint(100, 221, size=(30, 1))

# Preise generieren, basierend auf der Größe mit zufälliger Streuung
# Ein Basispreis pro Quadratmeter plus ein Zufallsanteil
y = 3000 * X.squeeze() + np.random.normal(0, 75000, size=30)

# Sicherstellen, dass die Preise im Bereich 250000 bis 750000 liegen
y = np.clip(y, 250000, 750000)


# KNN-Regressor mit k=3 Nachbarn erstellen
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X, y)

# Einen kontinuierlichen Bereich für Vorhersagen erstellen
X_plot = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
y_plot = knn.predict(X_plot)

# Plot erstellen
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color="black", label="Daten")
plt.plot(X_plot, y_plot, color="red", label="KNN-Vorhersage (k=3)")
plt.xlabel("Hausgröße (qm)")
plt.ylabel("Preis (EUR)")
plt.title("K-Nearest-Neighbors-Regression (1D-Beispiel)")
plt.legend()
plt.show()
```
