```python
import pandas as pd

def download_books():
  # Herunterladen des Datensatzes aus dem öffentlich verfügbaren Repository.
  books = pd.read_csv(
      "./Buchpreise.csv",
      sep=',')
  return books

books = download_books()
# Anzeigen der ersten Zeilen des Datensatzes.
books
```

```python
books = books.drop('ID', axis=1)
books.corr()
```

```python
# Löschen der Datensätze ohne Label
books = books[books['Preis'].notna()]
books
```

```python
from sklearn.impute import SimpleImputer

# Erstellen eines SimpleImputer-Objekts mit der Strategie 'mean'.
imputer = SimpleImputer(strategy='mean')

# Anwenden des Imputers auf die Spalte 'Blattdicke'.
books['Blattdicke'] = imputer.fit_transform(books[['Blattdicke']])
books
```

```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Anwenden des Iterativen Imputers
imputer = IterativeImputer()
books[['Anz. Autoren',	 'Seiten']] = imputer.fit_transform(books[['Anz. Autoren',	 'Seiten']])
books
```
