```python
import matplotlib.pyplot as plt

# Beispiel-Daten
klassen = ['Krieger', 'Magier', 'Schurke', 'Priester', 'Jger', 'Hexenmeister', 'Paladin', 'Druide', 'Todesritter', 'Mnch', 'Dmonenjger', 'Schamane', 'Jger', 'Hexenmeister', 'Paladin']
dps = [5000, 7000, 6500, 3000, 6000, 6800, 5500, 6200, 5700, 5900, 7100, 4800, 6000, 6800, 5500]
ueberlebensfaehigkeit = [80, 50, 60, 90, 70, 55, 85, 75, 65, 78, 52, 88, 70, 55, 85]


# Scatter-Plot erstellen
plt.figure(figsize=(10, 6))
plt.scatter(dps, ueberlebensfaehigkeit, color='blue')

# Beschriftungen hinzufgen
for i, klasse in enumerate(klassen):
    plt.text(dps[i] + 100, ueberlebensfaehigkeit[i], klasse, fontsize=9)

plt.xlabel('Schaden pro Sekunde (DPS)')
plt.ylabel('berlebensfhigkeit')
plt.title('DPS vs. berlebensfhigkeit im Gaming')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Daten der Tiere
tiere = {
    "Tier": ["Pfau", "Faultier", "Zwergkaninchen", "Koala", "Krake", "Igel", "Ente", "Panda", "Hyne", "Erdmnnchen",
             "Knguru", "Schildkrte", "Papagei", "Eichhrnchen", "Flamingo"],
    "Drama": [10, 1, 6, 3, 8, 5, 7, 4, 9, 8, 7, 2, 9, 5, 8],
    "Bewegung": [2, 1, 9, 2, 7, 4, 6, 3, 8, 9, 8, 1, 6, 7, 5]
}

# DataFrame erstellen
df = pd.DataFrame(tiere)

# K-Means-Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[['Drama', 'Bewegung']])

# Cluster-Namen zuweisen
cluster_namen = {
    0: "Drama-Queens im Chill-Modus",
    1: "Turbulente Theaterstars",
    2: "Sportliche Minimalisten",
    3: "Chillige Zeitgenossen"
}
df['Cluster_Name'] = df['Cluster'].map(cluster_namen)

# Scatterplot erstellen
plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green', 'purple']
for cluster, color in zip(range(4), colors):
    cluster_daten = df[df['Cluster'] == cluster]
    plt.scatter(cluster_daten['Drama'], cluster_daten['Bewegung'], color=color, label=cluster_namen[cluster], s=100)

# Tiernamen als Annotationen hinzufgen
for i, row in df.iterrows():
    plt.text(row['Drama'] + 0.1, row['Bewegung'] + 0.1, row['Tier'], fontsize=9)

# Plot-Einstellungen
plt.title("Tier-Clustering: Drama vs. Bewegung", fontsize=16)
plt.xlabel("Drama-Potenzial (110)", fontsize=12)
plt.ylabel("Bewegungsdrang (110)", fontsize=12)
plt.xticks(range(1, 11))
plt.yticks(range(1, 11))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title="Cluster", bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot anzeigen
plt.tight_layout()
plt.show()

# Cluster-Zuordnung ausgeben
print(df[['Tier', 'Cluster_Name']])
```

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans


# Scatterplot erstellen
plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green', 'purple']
for cluster, color in zip(range(4), colors):
    cluster_daten = df[df['Cluster'] == cluster]
    plt.scatter(cluster_daten['Drama'], cluster_daten['Bewegung'], color=color, label=cluster_namen[cluster], s=100)

# Tiernamen als Annotationen hinzufgen
for i, row in df.iterrows():
    plt.text(row['Drama'] + 0.1, row['Bewegung'] + 0.1, row['Tier'], fontsize=9)

# Zentroide hinzufgen
centroids = kmeans.cluster_centers_
for i, centroid in enumerate(centroids):
    plt.scatter(centroid[0], centroid[1], marker='x', s=200, color=colors[i], linewidths=3)
    plt.text(centroid[0] + 0.2, centroid[1] + 0.2, cluster_namen[i], fontsize=10, color=colors[i])


# Plot-Einstellungen
plt.title("Tier-Clustering: Drama vs. Bewegung", fontsize=16)
plt.xlabel("Drama-Potenzial (110)", fontsize=12)
plt.ylabel("Bewegungsdrang (110)", fontsize=12)
plt.xticks(range(1, 11))
plt.yticks(range(1, 11))
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title="Cluster", bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot anzeigen
plt.tight_layout()
plt.show()

# Cluster-Zuordnung ausgeben
print(df[['Tier', 'Cluster_Name']])
```
