```python
import Levenshtein

def find_best_match(query, candidates):
    """Findet die beste Übereinstimmung für eine Anfrage in einer Liste von Kandidaten."""
    best_score = 0
    best_match = None

    for candidate in candidates:
        # Berechne die Levenshtein-Ratio
        score = Levenshtein.ratio(query, candidate)

        if score > best_score:
            best_score = score
            best_match = candidate

    return best_match, best_score * 100.0

# Liste bekannter Straßen
known_streets = [
    "Hauptstraße", "Bahnhofstraße", "Schulstraße", "Gartenstraße",
    "Bergstraße", "Kirchstraße", "Waldstraße", "Dorfstraße",
    "Lindenstraße", "Birkenweg", "Eichenweg", "Tannenweg"
]

# Eingabe mit Tippfehlern
input = "Eichelweg" #"Huaptstr"

# Beste Übereinstimmung finden
best_match = find_best_match(input, known_streets)
print(f"Meintest du: {best_match[0]}? (Ähnlichkeit: {best_match[1]:.1f}%)")
```

```python
def normalize_german(text):
    """Normalisiert deutsche Schreibweisen für besseren Vergleich."""
    replacements = {
        'ä': 'ae', 'ö': 'oe', 'ü': 'ue', 'ß': 'ss',
        'Ä': 'Ae', 'Ö': 'Oe', 'Ü': 'Ue' }

    for old, new in replacements.items():
        text = text.replace(old, new)

    return text
```

```python
!pip install jellyfish
import jellyfish

# Deutsche Namensbeispiele
namen = [
    "Schmidt", "Schmitt", "Schmied", "Schmid",
    "Meier", "Meyer", "Maier", "Mayer",
    "Müller", "Mueller", "Miller",
    "Schäfer", "Schaefer", "Schaeffer"
]

# Ausgabe der Soundex-Codes
print("Soundex-Codes für deutsche Namen:")
for name in namen:
    print(f"{name}: {jellyfish.soundex(name)}")

# Gruppiere Namen mit gleichem Soundex-Code
soundex_gruppen = {}
for name in namen:
    code = jellyfish.soundex(name)
    if code not in soundex_gruppen:
        soundex_gruppen[code] = []
    soundex_gruppen[code].append(name)

print("\nNamen mit gleichem Klang (gleicher Soundex-Code):")
for code, gruppe in soundex_gruppen.items():
    if len(gruppe) > 1:  # Nur Gruppen mit mehreren Namen anzeigen
        print(f"Code {code}: {', '.join(gruppe)}")

# Beispiel für die Suche nach ähnlich klingenden Namen
suchname = "Maier"
soundex_code = jellyfish.soundex(suchname)
print(f"\nNamen, die ähnlich wie '{suchname}' klingen (Code: {soundex_code}):")
for name in namen:
    if jellyfish.soundex(name) == soundex_code and name != suchname:
        print(f"- {name}")
```
